```
import numpy as np
import pandas as pd
import pickle
import tweepy
from tweepy import OAuthHandler
import json
import jsonlines
```

###Providing twitter api access tokens

```
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''
```

```
# Setting authentication using api token keys
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
# Creating api object while calling auth information
api = tweepy.API(auth)
```

```
# Name of the twitter account
name = "midasIIITD"
#No. of tweets which can be accessed by api
tweetCount = 200
# Storing the result in results
results = api.user_timeline(id=name, count=tweetCount)
```

```
# Creating JSONlines file for accessing data
with jsonlines.open("midas_twitter.jsonl",mode='w') as writer:
    for tweet in results:
        writer.write(tweet._json)
```

```
# Function for checking the images
def check_images(entity):
  try:
   media = entity['media']
   return(len(media))
  except KeyError:
   return 'None'
```

```
data=[]
# Accessing the JSONline file for further information
with jsonlines.open("midas_twitter.jsonl",mode='r') as reader:
    for item in reader.iter():
        temp=[]
        temp.append(item['text'])
        temp.append(item['created_at'][:16])
        temp.append(item['favorite_count'])
        temp.append(item['retweet_count'])
#         Calling the check_images function for getting the count of images
        image_count = check_images(item['entities'])
        temp.append(image_count)
        data.append(temp)
        
df=pd.DataFrame(data,columns=['Text','Date and time','Number of likes','Nmber of retweets','Number of images'])
```
